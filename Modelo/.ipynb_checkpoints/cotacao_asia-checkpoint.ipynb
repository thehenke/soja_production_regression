```python
import pandas as pd
import numpy as np
from sklearn.model_selection import TimeSeriesSplit
from sklearn.linear_model import LinearRegression
import sklearn.metrics as metrics
```

```python
df_china = pd.read_csv('data/soja_cotacao_china_DSAc1.csv')
df_brasil = pd.read_csv('data/soja_cepea_brasil.csv', delimiter = ';')
df_china
```

```python
old_last = df_china['Último']
last = [a.split(',')[0] for a in old_last]

old_min = df_china['Último']
new_min = [a.split(',')[0] for a in old_min]

old_max = df_china['Último']
new_max = [a.split(',')[0] for a in old_max]
```

```python
df_china['Último'] = last
df_china['Máxima'] = new_max
df_china['Mínima'] = new_min
```

```python
df_china['data'] = df_china['Data'].str.replace('.','/')
df_china['Var%'] = df_china['Var%'].str.replace('%','')

df_china['data'] = pd.to_datetime(df_china['data'])
df_brasil['data'] = pd.to_datetime(df_brasil['data'])

df_china = df_china.set_index('data')
df_brasil = df_brasil.set_index('data')
```

```python
df_china = df_china.drop(['Data', 'Vol.', 'Abertura'], axis = 1)
```

```python
df_china
```

```python
df = pd.merge(df_china, df_brasil, left_index=True, right_index=True)
df.columns = ['last', 'max', 'min', 'var', 'price_br', 'price_us']
```

```python
df = df.astype(str).stack().str.replace(',','.').unstack().astype(float)
df
```
